```python
%matplotlib inline

import pandas as pd
import numpy as np
import pylab as plt

import portfolioopt as pfopt
from scipy.stats import norm

import seaborn as sns
import os

import sys
sys.path.append('C:\\projects\\finances.git')

from finances.market import market_data as mkt_data

data_path = 'C:\\projects\\finances.git\\finances\\trading\\strategies\\dynamic_stoploss\\data'
```

```python
def sharpe(data):
    return abs(data[data>=0].sum())/abs(data[data<0].sum())
    #return data.mean()/data.std()
```

# BTC

```python
df = pd.read_csv(os.path.join(data_path, 'dynamic_stoploss_strategy_BTC_Julho.csv'))

fig, ax = plt.subplots(1,1, figsize=(20,10))

df[df['period']=='12H'].boxplot(
    column='strategy',
    by=['period','pct_gap', 'min_gain'],
    rot=90,
    ax=ax)
plt.tight_layout()
plt.show()
```

```python
db = df.groupby(['period', 'pct_gap', 'min_gain']).agg(sharpe)
```

```python
fig, ax = plt.subplots(1,5, figsize=(15,8), sharey=True)
for n, period in enumerate(['4H', '6H', '8H', '12H', '24H']):
    b = db.unstack(['period', 'pct_gap'])
    b['strategy'][period].plot(style='o-', ax=ax[n], legend=(n<1))
    ax[n].set_title(period)
```

### Best performance:
* Period: 8H
* Pct Gap: 0.04
* Min gain: 0.02
